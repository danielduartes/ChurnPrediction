# Transformação de dados

A coluna 'TotalCharges' estava com alguns valores iguais a ' ' ou '', impossibilitando a transformação da coluna de string para float.

Com essas mudanças foi possível a percepção da existência de valores nulos e a transformação do tipo da coluna.

Essa transformação de ' ' para NaN foi realizada em todas as colunas para tratar possíveis inconsistêncas.

```python
df = df.replace(r'^\s*$', np.nan, regex=True)

for col in df.select_dtypes(include='object').columns:
    try:
        df[col] = pd.to_numeric(df[col])
    except:
        pass
```

## Verificando a frequência das colunas categóricas em relação ao target 'Churn'

Devido ao fato de das 21 colunas do dataframe, 17 serem categóricas (sem contar com a coluna do ID) não houve um grande leque de opções de gráficos a serem plotados

```python
fig, axes = plt.subplots(4, 4, figsize=(15, 15))
axes = axes.flatten()

colunas_sem_churn = [col for col in colunas_categoricas if col != 'Churn']

for i, col in enumerate(colunas_sem_churn):
    df.groupby(col)['Churn'].value_counts(normalize=True).unstack().plot(kind='bar', ax=axes[i])
    axes[i].set_title(f'{col} vs Churn')

plt.tight_layout()
plt.show()
```

## Feature engeneering

```python
# Transformando as colunas categóricas de dois valores em flag binária

df['gender'] = df['gender'].replace({'Female': 1, 'Male': 0})
df['SeniorCitizen'] = df['SeniorCitizen'].replace({'Yes': 1, 'No': 0})
df['PhoneService'] = df['PhoneService'].replace({'Yes': 1, 'No': 0})
df['Partner'] = df['Partner'].replace({'Yes': 1, 'No': 0})
df['Dependents'] = df['Dependents'].replace({'Yes': 1, 'No': 0})
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
```

```python
# As colunas categóricas que possuem mais de 2 valores possíveis é feito o 'get_dummies', em que cada valor ganha uma coluna própria que caso o valor seja True ele vira 1

colunas_categoricas_3_valores = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'TechSupport', 'StreamingTV',
                                  'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'DeviceProtection']

for coluna in colunas_categoricas_3_valores:
    df = pd.get_dummies(df, columns=[coluna], dtype=int)
```

```python
# A coluna 'customerID' foi dropada pelo fato de ser uma coluna utilizada apenas para identificação e não colaborar para a previsão do target
df = df.drop(columns='customerID')
```

Devido à grande quantidade de colunas após o 'get_dummies', para melhor visualização da correlação das colunas com o 'Churn' foi feito um gráfico de barras.

```python
correlacao_churn = df.corr()['Churn'].sort_values(ascending=False).drop('Churn')

plt.figure(figsize=(10, 12))
sns.barplot(x=correlacao_churn.values, y=correlacao_churn.index)
plt.axvline(x=0, color='black', linewidth=0.8)
plt.title('Correlação com Churn')
plt.tight_layout()
plt.show()
```

***Colunas com correlação entre -0.05 e 0.05:***
- PhoneService
- gender

Colunas com correlação entre -0.05 e 0.05 serão dropadas

```python
df = df.drop(columns=['PhoneService', 'gender'])
```
